```python
import os
from pathlib import Path
from collections import defaultdict

# Get the directory where the notebook is located
notebook_dir = Path.cwd()
data_dir = notebook_dir

# Dictionary to store counts for each CVE folder
before_file_counts = defaultdict(int)

# Iterate through each CVE folder
for cve_folder in sorted(data_dir.iterdir()):
    if cve_folder.is_dir() and cve_folder.name.startswith("CVE-"):
        # Count files with '_before_' in their name
        before_files = list(cve_folder.glob("*_before_*"))
        before_file_counts[cve_folder.name] = len(before_files)

# Display results
print(f"{'CVE Folder':<20} {'Before Files Count':<20}")
print("-" * 40)
for cve_name, count in before_file_counts.items():
    print(f"{cve_name:<20} {count:<20}")

print("\n" + "=" * 40)
print(f"Total CVE folders: {len(before_file_counts)}")
print(f"Total before files: {sum(before_file_counts.values())}")
print(f"Average before files per CVE: {sum(before_file_counts.values()) / len(before_file_counts):.2f}")

# Return the counts dictionary
before_file_counts
```
